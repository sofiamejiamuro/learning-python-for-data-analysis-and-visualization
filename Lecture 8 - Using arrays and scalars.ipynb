## MULTIPLICATION

The process of matrix multiplication using NumPy is known as **vectorization**.The main objective of vectorization is to remove or reduce the for loops which we were using explicitly. By reducing 'for' loops from programs gives faster computation. The build-in package NumPy is used for manipulation and array-processing.
